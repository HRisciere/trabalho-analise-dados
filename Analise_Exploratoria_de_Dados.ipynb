```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

<h1>Configuração Gerais</h1>

```python
# Configurações de visualização
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")
```

<h1>Importação Dados</h1>

```python
train = pd.read_csv('Data/train.csv')
test = pd.read_csv('Data/test.csv')
sample_submission = pd.read_csv('Data/sample_submission.csv')
```

```python
print("Train:")
display(train.head())

print("Test:")
display(test.head())

print("Sample Submission:")
display(sample_submission.head())
```

# Verificando as Colunas e os Tipos de Dados

```python
def resumo_tipos(df, nome):
    print(f"\n{nome.upper()}:")
    print(f"Formato: {df.shape[0]} linhas e {df.shape[1]} colunas\n")
    
    cat = df.select_dtypes(include='object').columns.tolist()
    num = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

    print(f"Variáveis categóricas ({len(cat)}): {cat}")
    print(f"\nVariáveis numéricas ({len(num)}): {num}")

resumo_tipos(train, 'train')
resumo_tipos(test, 'test')
resumo_tipos(sample_submission, 'sample_submission')
```

# Verificando Valores Faltantes

```python
def valores_faltantes(df, nome):
    faltantes = df.isnull().sum()
    faltantes = faltantes[faltantes > 0].sort_values(ascending=False)
    print(f"\n{nome.upper()} - Total de colunas com valores faltantes: {faltantes.shape[0]}")
    display(faltantes.to_frame('Total Faltante'))

valores_faltantes(train, 'train')
valores_faltantes(test, 'test')
valores_faltantes(sample_submission, 'sample_submission')
```

# Distribuição da Variável Alvo (SalePrice)

```python
plt.figure(figsize=(10,6))
sns.histplot(train['SalePrice'], kde=True, color='blue')
plt.title('Distribuição de SalePrice')
plt.xlabel('SalePrice')
plt.ylabel('Frequência')
plt.show()

print(f"Assimetria (Skewness) da variável SalePrice: {train['SalePrice'].skew():.2f}")
```

<p>A distribuição apresenta uma assimetria positiva (Skewness = 1.88), ou seja, a cauda está puxada para a direita.</p>
<p>Isso significa que existem algumas casas com preços muito mais altos do que a maioria, o que pode impactar modelos sensíveis à normalidade, como Regressão Linear.</p>

# Análise de Correlação (Matriz de Correlação)

```python
dados_numericos = train.select_dtypes(include=['int64', 'float64'])

corr = dados_numericos.corr()

top_corr = corr['SalePrice'].abs().sort_values(ascending=False).head(11).index

plt.figure(figsize=(10,8))
sns.heatmap(dados_numericos[top_corr].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlação das Principais Variáveis com SalePrice')
plt.show()

corr_SalePrice = corr['SalePrice'].sort_values(ascending=False)
display(corr_SalePrice)
```

<p> Alta correlação:
    <ul>
        <li>OverallQual e GrLivArea são os principais drivers do preço de venda das casas.</li>
        <li>As variáveis relacionadas à garagem (GarageCars, GarageArea) também são bem relevantes.</li>
        <li>Tamanho de porão (TotalBsmtSF) e primeiro andar (1stFlrSF) são importantes.</li>
    </ul>
</p> 
<p> Baixa correlação::
    <ul>
        <li>Variáveis como KitchenAbvGr, EnclosedPorch, YrSold e OverallCond têm correlação baixa ou negativa.</li>
    </ul>
</p> 

# Relação entre as principais variáveis e SalePrice

```python
# Selecionando algumas variáveis altamente correlacionadas
variaveis = ['GrLivArea', 'OverallQual', 'TotalBsmtSF', 'GarageArea', '1stFlrSF', 'YearBuilt']

plt.figure(figsize=(18, 12))
for i, var in enumerate(variaveis):
    plt.subplot(2, 3, i + 1)
    sns.scatterplot(x=train[var], y=train['SalePrice'])
    plt.title(f'SalePrice x {var}')
plt.tight_layout()
plt.show()
```

Análise das Relações com SalePrice

1. GrLivArea x SalePrice
-Relação linear clara e forte.
-Quanto maior a área construída acima do solo, maior tende a ser o preço.
-Existem alguns outliers, como casas com alta área mas preço relativamente baixo.

2. OverallQual x SalePrice
-Relação fortemente positiva e quase discreta, já que OverallQual é uma variável ordinal (de 1 a 10).
-Aumentos na qualidade constroem saltos claros no preço.
-Uma das variáveis mais importantes para precificação.

3. TotalBsmtSF x SalePrice
-Relação positiva, mas não tão forte quanto GrLivArea.
-Algumas casas com grandes áreas de porão não possuem preços tão altos — isso sugere que só o tamanho do porão não é tão decisivo isoladamente.

4. GarageArea x SalePrice
-Existe relação, porém com muitos pontos agrupados e dispersão maior.
-Mostra que garagem influencia, mas não é determinante sozinha.

5. 1stFlrSF x SalePrice
-Relação similar a GrLivArea, porém um pouco mais dispersa.
-Continua sendo uma variável relevante, mas provavelmente mais fraca que a soma dos andares (GrLivArea).

6. YearBuilt x SalePrice
-Tendência clara: casas mais novas possuem valores maiores.
-A partir dos anos 1980-1990 há um aumento evidente nos preços.
-Construções anteriores a 1950 tendem a ter preços bem mais baixos, reforçando o impacto da modernidade.
